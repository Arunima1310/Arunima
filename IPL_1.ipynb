---------plots--------

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(p['season'], p['win_by_runs'])
plt.xlabel('Season')
plt.ylabel('Win by Runs')
plt.title('Scatter Plot of Season vs. Win by Runs')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.countplot(x='winner', data=p, order=p['winner'].value_counts().index)
plt.xlabel('Winning Team')
plt.ylabel('Number of Wins')
plt.title('Number of Wins per Team')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.boxplot(x='venue', y='win_by_runs', data=p)
plt.xlabel('Venue')
plt.ylabel('Win by Runs')
plt.title('Distribution of Win by Runs for Different Venues')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='toss_decision', y='win_by_wickets', data=p)
plt.xlabel('Toss Decision')
plt.ylabel('Win by Wickets')
plt.title('Distribution of Win by Wickets for Different Toss Decisions')
plt.show()
```

```python
import seaborn as sns

numericcolumns = ['season', 'win_by_runs', 'win_by_wickets']
correlation = p[numericcolumns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
wins_venue_team = p.groupby(['venue', 'winner']).size().unstack(fill_value=0)
wins_venue_team.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Venue')
plt.ylabel('Number of Wins')
plt.title('Wins by Venue and Team')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='city', data=p, order=p['city'].value_counts().index)
plt.xlabel('Number of Matches')
plt.ylabel('City')
plt.title('Number of Matches Played in Each City')
plt.show()
```

```python
toss_decisions = p['toss_decision'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(toss_decisions, labels=toss_decisions.index, autopct='%1.1f%%', startangle=90)
plt.title('Toss Decisions')
plt.show()
```

```python
import matplotlib.pyplot as plt
winbyseason_team = p.groupby(['season', 'winner']).size().unstack(fill_value=0)
plt.figure(figsize=(20, 12))
for team in winbyseason_team.columns:
    plt.plot(winbyseason_team.index, winbyseason_team[team], label=team)
plt.ylabel('Team')
plt.xlabel('Season')
plt.title('Wins by Season for Each Team')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
